x1 = 0 -> 구름 0<br>
x1 = 1 -> 구름 100%<br>

x2 = 0 -> 바람 0<br>
x2 = 1 -> 바람 100%<br>

y = 0 -> 맑음<br>
y = 1 -> 비<br>


```python
import random

w1 = random.random()
w2 = random.random()
```

```python
x1 = [0.8,0.6,0.1,0.3,0.6,0.4,0.1]
x2 = [0.6,0.9,0.2,0.1,0.6,0.3,0.2]

y = [1,1,0,0,1,0,0]
```

```python
# 가중합
def weightedSum(x1_v: float, x2_v: float) -> float:
    return x1_v * w1 + x2_v * w2

# step activate 함수
def activationFunc(weighted_sum: float) -> int:
    if weighted_sum < 0.5:
        return 0
    else:
        return 1

for k in range(30):
    for i in range(len(x1)):
        sum_value = weightedSum(x1[i], x2[i])
        predict = activationFunc(sum_value)

        if predict == y[i]:
            error = 0
        else:
            error = 1

        learningRate = 0.01
        w1 += error * x1[i] * learningRate
        w2 += error * x2[i] * learningRate

        print(w1, w2)

```

```python
for i in range(len(x1)):
  x1_v = x1[i]
  x2_v = x2[i]
  weighted = x1_v * w1 + x2_v * w2

  if weighted < 0.5:
    print(f"{x1_v, x2_v} -> 0, real: {y[i]}")
  else: 
    print(f"{x1_v, x2_v} -> 1, real: {y[i]}")
```
