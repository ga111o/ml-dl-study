```python
import random

X = [[0, 0], [0, 1], [1, 0], [1, 1]] 
Y = [0, 0, 0, 1]



def predict(X, weights, bias):
  output = 0

  for i in range(len(weights)):
    output += weights[i] * X[i]

  output += bias

  if output >= 0:
    return 1
  else:
    return 0


def learning(X, Y, learning_rate, iter_num):
  input_size = len(X[0])
  weights = []

  for i in range(input_size):
    weights.append(random.random())

  bias = random.random()

  for i in range(iter_num):
    for input_sample, target_value in zip(X, Y):
  
      output = predict(input_sample, weights, bias)
      
      error = target_value - output
      update = learning_rate * error

      for index in range(input_size):
        weights[index] += update * input_sample[index]
  
      bias += update 

  return weights, bias

weights, bias = learning(X, Y, 0.1, 10)

for x_i in X:
  print(f"input: {x_i}, predict: {predict(x_i, weights, bias)}")
```
