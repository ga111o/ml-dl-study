## bayes classifier

risk: expectated loss

데이터 포인트가 특정 클래스에 속할 확률을 계산하는 classifier




lisk를 minimazation해주는 function = f\* = bayes classifier


$
P(Y|X) = \frac{P(X|Y) \cdot P(Y)}{P(X)}
$

$ P(Y|X) $ : likelyhood, X가 Y에 속할 확률

$ P(Y) $ : 전체 경우 중 Y가 나올 확률 ...... (1, 0을 classification 할 경우 -> 1이 나올 확률 70일 때 P(1) = 70)

$ P(X) $ : 클래스에 대해 X가 나올 확률을 P(Y)로 가중 평균한 값, 서로 상쇄돼서 없어지기에 크게 상관 없음

<br>
