```python
import random
import numpy as np

def getTrainingData(data_point):
  x1 = np.random.random(data_point)
  x2 = np.random.random(data_point)
  y = ((x1+x2)>1).astype(int)
  train_set = [((x1[i],x2[i]), y[i]) for i in range(len(x1))]

  return train_set

data_set = getTrainingData(300)

w1 = random.random()
w2 = random.random()

# 가중합
def weightedSum(x1_v: float, x2_v: float) -> float:
    return x1_v * w1 + x2_v * w2

# step activate 함수
def activationFunc(weighted_sum: float) -> int:
    if weighted_sum < 0.5:
        return 0
    else:
        return 1

for k in range(30):
    for i in range(len(data_set)):
        sum_value = weightedSum(data_set[i][0][0], data_set[i][0][1])
        predict = activationFunc(sum_value)

        error = data_set[i][1] - predict
        print(f"error: {error} | w1: {w1} | w2: {w2}")

        learningRate = 0.01
        w1 += error * data_set[i][0][0] * learningRate
        w2 += error * data_set[i][0][1] * learningRate

        # print(w1, w2)

```
